## 🎯 성공 기준
- [ ] `await agent.create_session()`으로 세션 생성
- [ ] 모든 `run()` 호출에 `session=` 전달
- [ ] 3번째 대화에서 첫 번째에 말한 이름/과목을 정확히 기억

## 💡 MAF Session 핵심 패턴
```python
# 세션 생성 (대화 이력 자동 관리)
session = await agent.create_session()

# 같은 session 전달 → 이전 대화 기억!
r1 = await agent.run("내 이름은 Alice야", session=session)
r2 = await agent.run("내 이름이 뭐야?",   session=session)  # ✅ 기억함!
```
