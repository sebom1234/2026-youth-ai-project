# 🚀 프롬프트 엔지니어링으로 AI와 잘 대화하기

> AI 입문자를 위해, 좋은 프롬프트가 왜 좋은 답변을 만드는지 **직접 비교하고 실험하는** 노트북입니다.

---

## ChatGPT에게 어떻게 말해야 좋은 답을 받을까?

AI는 마음을 읽지 못합니다. 사용자가 적어준 정보와 지시를 바탕으로 다음 답변을 만듭니다.

```
나쁜 요청: "공부법 알려줘"
좋은 요청: "고등학교 2학년이 4주 동안 영어 단어를 외울 수 있는 계획을 표로 만들어줘. 하루 30분 기준으로 해줘."
```

이 노트북에서는 좋은 프롬프트를 만드는 방법을 단계별로 연습합니다.
- 모호한 요청을 구체적인 요청으로 바꾸기
- 역할, 목표, 맥락, 형식, 제약 넣기
- 답변을 보고 다시 개선하기

**핵심 메시지**: *프롬프트 엔지니어링은 AI에게 일을 잘 맡기기 위한 명확한 의사소통 기술입니다.*

## 1단계: 문제 상황 — 모호한 요청은 모호한 답을 만든다

친구에게 부탁할 때도 "도와줘"라고만 말하면 무엇을 어떻게 도와야 할지 알기 어렵습니다. AI도 같습니다.

좋은 프롬프트는 보통 다음 정보를 더 많이 담고 있습니다.
- 내가 원하는 목표
- 현재 상황과 배경
- 답변을 받을 형식
- 지켜야 할 조건
- 답변의 수준과 말투

먼저 같은 주제의 짧은 프롬프트와 구체적인 프롬프트를 비교해봅니다.

````python
examples = [
    {
        "situation": "영어 공부",
        "basic": "영어 공부법 알려줘",
        "better": "고등학교 2학년이 토익 600점을 목표로 3개월 동안 공부할 수 있는 주별 계획을 표로 만들어줘. 하루 공부 시간은 1시간으로 가정해줘.",
    },
    {
        "situation": "글쓰기",
        "basic": "소설 써줘",
        "better": "고등학교를 배경으로 한 청춘 단편소설의 도입부를 800자로 써줘. 주인공은 전학생이고, 분위기는 처음엔 어색하지만 따뜻하게 마무리되게 해줘.",
    },
    {
        "situation": "진로 상담",
        "basic": "진로 추천해줘",
        "better": "수학을 좋아하고 사람들과 협업하는 것을 즐기는 고등학생에게 어울리는 진로 5가지를 추천해줘. 각 진로마다 필요한 역량과 준비 방법도 알려줘.",
    },
]

print("📌 기본 프롬프트 vs 개선된 프롬프트")
for item in examples:
    print(f"\n[{item['situation']}]")
    print(f"기본: {item['basic']}")
    print(f"개선: {item['better']}")
````

## 2단계: 프롬프트의 재료 — 역할, 목표, 맥락, 형식, 제약

좋은 프롬프트를 만드는 기본 재료는 다섯 가지로 기억하면 쉽습니다.

| 재료 | 뜻 | 예시 |
|---|---|---|
| 역할 | AI가 어떤 입장으로 답할지 | "너는 친절한 수학 선생님이야" |
| 목표 | 무엇을 해주길 원하는지 | "이차방정식을 설명해줘" |
| 맥락 | 사용자의 상황과 배경 | "처음 배우는 고등학생 기준" |
| 형식 | 답변 모양 | "표와 예시 2개로 정리" |
| 제약 | 지켜야 할 조건 | "500자 이내, 쉬운 말로" |

이 다섯 가지를 모두 넣으면 AI가 훨씬 안정적으로 답할 수 있습니다.

````python
def build_prompt(role, goal, context, output_format, constraints):
    return f"""역할: {role}
목표: {goal}
맥락: {context}
답변 형식: {output_format}
제약 조건: {constraints}"""

prompt = build_prompt(
    role="친절한 고등학교 과학 선생님",
    goal="뉴턴의 운동 법칙 3가지를 설명하기",
    context="물리를 처음 배우는 고등학생이 이해할 수 있어야 함",
    output_format="각 법칙마다 쉬운 설명, 일상 예시 1개, 한 줄 요약",
    constraints="어려운 수식은 최소화하고 전체 답변은 700자 이내",
)

print("🧱 프롬프트 템플릿 예시")
print(prompt)
````

## 3단계: 좋은 프롬프트의 패턴 — 구체성이 답변 품질을 바꾼다

좋은 프롬프트에는 반복되는 패턴이 있습니다.

1. 무엇을 원하는지 분명하다
2. 대상 독자나 수준이 정해져 있다
3. 결과물의 형식이 정해져 있다
4. 길이, 난이도, 조건이 있다
5. 예시나 기준이 있다

이번에는 간단한 규칙으로 프롬프트가 얼마나 구체적인지 점수를 매겨봅니다. 실제 AI 평가가 아니라, 연습용 체크 도구입니다.

````python
check_keywords = {
    "역할": ["너는", "선생님", "전문가", "상담사", "코치"],
    "목표": ["설명", "추천", "작성", "분석", "계획", "비교"],
    "맥락": ["고등학생", "초보자", "처음", "상황", "배경", "목표"],
    "형식": ["표", "목록", "단계", "예시", "요약", "템플릿"],
    "제약": ["이내", "분량", "조건", "반드시", "하지 말고", "기준"],
}


def prompt_score(text):
    found = []
    for category, keywords in check_keywords.items():
        if any(keyword in text for keyword in keywords):
            found.append(category)
    score = len(found)
    return score, found

prompts = [
    "AI 설명해줘",
    "고등학생도 이해할 수 있게 AI의 기본 원리를 일상 예시 3개와 표로 설명해줘. 어려운 용어는 쉽게 풀어줘.",
]

for text in prompts:
    score, found = prompt_score(text)
    print("프롬프트:", text)
    print(f"점수: {score}/5 | 포함된 요소: {', '.join(found) if found else '없음'}")
    print()
````

## 4단계: 핵심 기법 — 상황에 따라 다른 방법 쓰기

프롬프트 엔지니어링에는 자주 쓰는 기법들이 있습니다.

| 기법 | 언제 쓰나 | 핵심 문장 |
|---|---|---|
| 역할 부여 | 답변 관점을 정하고 싶을 때 | "너는 ... 전문가야" |
| Zero-shot | 예시 없이 바로 답을 받고 싶을 때 | "예시 없이 바로 분류해줘" |
| Few-shot | 예시처럼 답하게 하고 싶을 때 | "예시 2개를 참고해서" |
| Chain-of-Thought(단계별 풀이) | 복잡한 문제를 풀 때 | "풀이 단계를 보여줘" |
| 형식 지정 | 결과를 바로 쓰고 싶을 때 | "표로 정리해줘" |
| 제약 조건 | 범위를 조절하고 싶을 때 | "500자 이내로" |
| 개선 요청 | 답변을 다듬고 싶을 때 | "더 쉽게 다시 써줘" |

### Zero-shot, Few-shot, Chain-of-Thought

- **Zero-shot**: 예시를 주지 않고 바로 요청합니다. 빠르게 초안을 만들거나 간단한 분류를 할 때 좋습니다.
- **Few-shot**: 원하는 답변 예시를 1~3개 보여준 뒤 같은 형식으로 새 답을 만들게 합니다. 출력 형식과 톤을 맞추는 데 좋습니다.
- **Chain-of-Thought**: 복잡한 문제에서 바로 정답만 받지 않고, 필요한 근거와 풀이 단계를 나누어 설명하게 합니다. 실제 사용에서는 AI의 숨은 사고과정을 요구하기보다, 검토 가능한 풀이 단계와 요약을 요청하는 것이 좋습니다.

중요한 것은 기법 이름을 외우는 것보다, 상황에 맞게 조합하는 것입니다.

````python
techniques = {
    "역할 부여": "너는 친절한 진로 상담 선생님이야. 고등학생 눈높이로 답해줘.",
    "Zero-shot": "예시 없이 바로 요청: 다음 문장을 긍정/부정/중립 중 하나로 분류해줘. 문장: 오늘 수업이 정말 재미있었다.",
    "Few-shot": "아래 예시 형식을 따라 새로운 예시 3개를 만들어줘. 예시: 질문 → 핵심 답 → 쉬운 비유",
    "Chain-of-Thought": "정답만 말하지 말고, 필요한 개념과 풀이 단계를 3단계로 나누어 설명한 뒤 마지막에 답을 요약해줘.",
    "형식 지정": "답변을 표로 정리하고, 마지막에 한 줄 요약을 붙여줘.",
    "제약 조건": "전체 답변은 600자 이내로 하고, 어려운 전문 용어는 사용하지 마.",
    "개선 요청": "방금 답변을 더 짧고 쉽게 바꿔줘. 중학생도 이해할 수 있게 예시를 추가해줘.",
}

print("🛠️ 바로 가져다 쓸 수 있는 기법 문장")
for name, sentence in techniques.items():
    print(f"\n[{name}]")
    print(sentence)
````

## 5단계: 반복 개선 — 첫 답변에서 끝내지 않기

좋은 결과는 보통 한 번에 나오지 않습니다. 첫 답변을 보고 다시 요청하면서 점점 원하는 방향에 가까워집니다.

예를 들어 이렇게 이어갈 수 있습니다.

1. 처음 요청하기
2. 답변에서 부족한 점 찾기
3. 더 구체적으로 다시 요청하기
4. 형식이나 수준을 조정하기

AI와의 대화는 검색창에 한 번 입력하고 끝내는 일이 아니라, 함께 초안을 다듬는 과정에 가깝습니다.

````python
def refine_prompt(original, feedback):
    return f"""기존 요청:
{original}

개선 요청:
{feedback}

위 피드백을 반영해서 더 좋은 답변을 다시 작성해줘."""

original_prompt = "시험 공부 계획 짜줘"
feedback = "고등학교 1학년, 중간고사 2주 전, 하루 2시간 공부 가능, 국어/수학/영어 중심으로 표를 원해."

print("🔄 개선된 후속 프롬프트")
print(refine_prompt(original_prompt, feedback))
````

## 6단계: 안전하고 책임 있게 쓰기 — AI 답변은 확인해야 한다

프롬프트를 잘 써도 AI가 항상 맞는 답을 주는 것은 아닙니다. 특히 중요한 정보는 반드시 확인해야 합니다.

주의할 상황:
- 의료, 법률, 금융처럼 실제 피해가 생길 수 있는 조언
- 최신 정보가 필요한 질문
- 출처가 중요한 과제나 보고서
- 개인정보가 포함된 내용

좋은 프롬프트는 단지 답을 잘 받는 기술이 아니라, AI를 책임 있게 쓰는 습관까지 포함합니다.

````python
safety_checklist = [
    "개인정보를 넣지 않았는가?",
    "최신 정보가 필요한 내용이면 확인 요청을 했는가?",
    "중요한 결정에 바로 사용하지 않고 검토할 계획이 있는가?",
    "출처나 근거가 필요한 경우 요청했는가?",
]

print("🧭 AI 사용 전 체크리스트")
for i, item in enumerate(safety_checklist, 1):
    print(f"{i}. {item}")
````

## 7단계: 실습 워크숍 — 나만의 프롬프트 만들기

아래 빈칸을 채운다고 생각하면서 나만의 프롬프트를 만들어보세요.

```text
너는 [역할]이야.
나는 [상황/배경]에 있어.
목표는 [원하는 결과]야.
답변은 [형식]으로 작성해줘.
조건은 [제약 조건]이야.
마지막에는 [검토 기준/요약]을 붙여줘.
```

추천 실습 주제:
- 수학 개념 설명 요청
- 독서 감상문 초안 만들기
- 진로 탐색 질문 만들기
- 수행평가 발표 자료 구성하기
- 하루 공부 계획 세우기

````python
my_prompt = build_prompt(
    role="꼼꼼한 발표 코치",
    goal="AI 윤리를 주제로 5분 발표문 구성하기",
    context="고등학교 정보 수업 수행평가이며, 청중은 같은 반 친구들",
    output_format="도입-본론-결론 구조와 슬라이드 5장 구성안",
    constraints="쉬운 예시를 포함하고, 발표 시간이 5분을 넘지 않게 구성",
)

score, found = prompt_score(my_prompt)
print("✍️ 내가 만든 프롬프트")
print(my_prompt)
print(f"\n체크 점수: {score}/5 | 포함된 요소: {', '.join(found)}")
````

## 📝 정리: 좋은 프롬프트에서 좋은 협업까지

| 항목 | 나쁜 프롬프트 | 좋은 프롬프트 |
|---|---|---|
| 목표 | 모호함 | 원하는 결과가 분명함 |
| 맥락 | 거의 없음 | 상황과 대상이 있음 |
| 형식 | AI가 알아서 정함 | 표, 목록, 단계 등 지정 |
| 제약 | 없음 | 길이, 수준, 조건이 있음 |
| 개선 | 한 번에 끝냄 | 답변을 보고 다시 다듬음 |

전체 흐름은 이렇게 기억하면 됩니다.

1. 목표 정하기 → 2. 맥락 넣기 → 3. 형식 정하기 → 4. 제약 넣기 → 5. 답변 검토하기 → 6. 다시 개선하기

> AI는 강력한 도구입니다. 좋은 프롬프트는 그 도구를 더 정확하고 책임 있게 쓰게 해주는 사용 설명서입니다.
